```python
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family = 'Malgun Gothic')

font1 = {'size' : 18, 'color' : 'blue'}

xdata = ['안지영', '홍지수', '황예린']
ydata1 = [90, 85, 88]
ydata2 = [83, 88, 91]
loc_list = [
    'upper left', 'upper center', 'upper right',
    'center left', 'center', 'center right',
    'lower left', 'lower center', 'lower right',
    'best'
]

plt.plot(xdata, ydata1, label = '국어')
plt.plot(xdata, ydata2, label = '영어')
plt.legend(loc = loc_list[9])
plt.title('국어, 영어 성적')
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family = 'Malgun Gothic')

font1 = {'size' : 18, 'color' : 'magenta'}

xdata = ['안지영', '홍지수', '황예린']
ydata1 = [90, 85, 88]
ydata2 = [83, 88, 95]
ydata3 = [85, 95, 78]
ydata4 = [92, 95, 82]
ydata5 = [95, 92, 85]

plt.plot(xdata, ydata1, label = '국어', color = 'red', linestyle = '-', marker = 'o')
plt.plot(xdata, ydata2, label = '영어', color = '#FFA500', linestyle = '--', marker = 'x')
plt.plot(xdata, ydata3, label = '수학', color = '#FFD700', linestyle = '-.', marker = 's')
plt.plot(xdata, ydata4, label = '물리', color = '#008000', linestyle = ':', marker = 'd')
plt.plot(xdata, ydata5, label = '화학', color = 'blue', linestyle = '-', marker = '*')

plt.title('성적표', fontdict = font1)
plt.legend(loc = 'best')

plt.show()
```

```python
import matplotlib.pyplot as plt

xdata = list(range(30))
ydata = []

for x in xdata :
    y = 2 * x
    ydata.append(y)

plt.plot(xdata, ydata, label = 'y = 2x')
plt.xlim(0, 35)
plt.ylim(0, 70)
plt.xticks(list(range(0, 35, 2)))
plt.yticks(list(range(0, 70, 5)))

plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family = 'Malgun Gothic')

pets = ['개', '고양이', '기타', '기르지않음']
portion = [18, 3.4, 3.1, 75.5]

plt.pie(portion, explode = (0, 0.1, 0, 0), labels = pets, autopct = '%1.1f%%', shadow = True, startangle = 90)
plt.title('총 가구 대비 반려동물 기르는 비율')

plt.show()
```
