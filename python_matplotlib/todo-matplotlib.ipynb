```python
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family = 'Malgun Gothic')

font1 = {'size' : 18, 'color' : 'blue'}

xdata = ['', '', '']
ydata1 = [90, 85, 88]
ydata2 = [83, 88, 91]
loc_list = [
    'upper left', 'upper center', 'upper right',
    'center left', 'center', 'center right',
    'lower left', 'lower center', 'lower right',
    'best'
]

plt.plot(xdata, ydata1, label = '')
plt.plot(xdata, ydata2, label = '')
plt.legend(loc = loc_list[9])
plt.title(',  ')
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family = 'Malgun Gothic')

font1 = {'size' : 18, 'color' : 'magenta'}

xdata = ['', '', '']
ydata1 = [90, 85, 88]
ydata2 = [83, 88, 95]
ydata3 = [85, 95, 78]
ydata4 = [92, 95, 82]
ydata5 = [95, 92, 85]

plt.plot(xdata, ydata1, label = '', color = 'red', linestyle = '-', marker = 'o')
plt.plot(xdata, ydata2, label = '', color = '#FFA500', linestyle = '--', marker = 'x')
plt.plot(xdata, ydata3, label = '', color = '#FFD700', linestyle = '-.', marker = 's')
plt.plot(xdata, ydata4, label = '', color = '#008000', linestyle = ':', marker = 'd')
plt.plot(xdata, ydata5, label = '', color = 'blue', linestyle = '-', marker = '*')

plt.title('', fontdict = font1)
plt.legend(loc = 'best')

plt.show()
```

```python
import matplotlib.pyplot as plt

xdata = list(range(30))
ydata = []

for x in xdata :
    y = 2 * x
    ydata.append(y)

plt.plot(xdata, ydata, label = 'y = 2x')
plt.xlim(0, 35)
plt.ylim(0, 70)
plt.xticks(list(range(0, 35, 2)))
plt.yticks(list(range(0, 70, 5)))

plt.grid(True)
plt.show()
```
