```python
# matplotlib에서 주석을 한글로 표기하기
import matplotlib.pyplot as plt

plt.tight_layout()
plt.axis([0, 100, 0, 100])
plt.plot(20, 20, 'bs')
plt.plot(50, 50, 'bs')
plt.plot(80, 80, 'bs')
plt.annotate('중앙값', xy=(50, 50), xytext=(50,20), fontsize=14, ha='center', 
    arrowprops=dict(facecolor='black', width=1, shrink=0.1, headwidth=10))
plt.text(25, 80, '상자 속에 글자', fontsize=24, bbox=dict(boxstyle='square', color='lightgray'))

plt.show()
```

```python
# font_manager 불러오기
import matplotlib.pyplot as plt
from matplotlib import font_manager

# 실제 경로로 가서 폰트 확인 해주세요
font_name = 'C:/Windows/Fonts/malgun.ttf'
font_family = font_manager.FontProperties(fname=font_name).get_name()
print(font_family)
```

```python
# 차트 font.family 속성을 font_family 변수로 가져와서 한글 적용
plt.rcParams['font.family'] = font_family

# 위에 내용을 복사
plt.tight_layout()
plt.axis([0, 100, 0, 100])
plt.plot(20, 20, 'bs')
plt.plot(50, 50, 'bs')
plt.plot(80, 80, 'bs')
plt.annotate('중앙값', xy=(50, 50), xytext=(50,20), fontsize=14, ha='center', 
    arrowprops=dict(facecolor='black', width=1, shrink=0.1, headwidth=10))
plt.text(25, 80, '상자 속에 글자', fontsize=24, bbox=dict(boxstyle='square', color='lightgray'))

plt.show()
```

```python
# matplot.annotate(화살표, 그리드, 텍스트 그리기)
import numpy as np
import matplotlib.pyplot as plt

plt.axis([-10, 10, -10, 10])
plt.vlines([0, 5, -5], -10, 10, linestyles=':') # 수직으로 점(0, 5, -5)에 -10에서 10까지 그리기
plt.hlines([0, 5 ,-5], -10, 10, linestyles=':') # ([그리는 좌표], 길이, 선 스타일)
```

```python
# 글 쓰기
plt.text(0, 0, 'hello', fontsize=20, va='top', ha='right')

# 위에 내용을 복사
plt.axis([-10, 10, -10, 10])
plt.vlines([0, 5, -5], -10, 10, linestyles=':') # 수직으로 점(0, 5, -5)에 -10에서 10까지 그리기
plt.hlines([0, 5 ,-5], -10, 10, linestyles=':') # ([그리는 좌표], 길이, 선 스타일)
```

```python
# 화살표 plt.annotate(문자, xy=화살표 방향, xytext=문자, 화살표 위치, arrowprops=화살표 성격)
plt.annotate('zero', xy=(0, 0), xytext=(5, 5), arrowprops=dict(facecolor='black')) 

# 글 쓰기
plt.text(0, 0, 'hello', fontsize=20, va='top', ha='right')

# 위에 내용을 복사
plt.axis([-10, 10, -10, 10])
plt.vlines([0, 5, -5], -10, 10, linestyles=':') # 수직으로 점(0, 5, -5)에 -10에서 10까지 그리기
plt.hlines([0, 5 ,-5], -10, 10, linestyles=':') # ([그리는 좌표], 길이, 선 스타일)
```
