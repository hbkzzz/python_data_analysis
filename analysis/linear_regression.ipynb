```python
# 선형 회귀 분석 회귀직선의 기울기와 절편 구하는법
# 예시 : 기온이 높아지면 주스의 판매량이 늘어나는지 분석

# 주피터 노트북 안에 산포도 표시
%matplotlib inline
# 연결
import matplotlib.pyplot as plt
import numpy as np

# 데이터
x = np.array([23, 24, 28, 24, 27, 21, 18, 25, 28, 20])   # 기온
y = np.array([37, 22, 62, 32, 74, 16, 10, 69, 83, 7])    # 주스의 판매량

# 회귀직선
a, b = np.polyfit(x, y, 1)
y2 = a * x + b
print('기울기: {0}, 절편: {1}'.format(a, b))    # 기울기와 절편을 표시

# 그리기
plt.scatter(x, y)   # 산포도
plt.plot(x, y2)     # 회귀직선
plt.show()
```
